Piezoelectric sensors

- The value of g33 typically varies from 0.019-0.0306 Vm/N based on the type of material. 
https://info.piezo.com/hubfs/Data-Sheets/piezo-material-properties-data-sheet-20201112.pdf
- Assuming a piezoelectric voltage constant of 0.020, a 0.1 mm distance between the plates and an area of 3.14*(0.007 meters)^2 (
    https://www.murata.com/en-us/api/pdfdownloadapi?cate=cgsubDiaphragms&partno=7BB-20-6CL0
) = will generate a voltage of V = (g33)Fd/A = (0.020*0.0001)*F/(3.14*(0.007)^2) = 0.0129988301 N. This means that for our desired voltage of 2V, we would need a force of 2/0.0129988301 = 153.86N (34.59 lbs)
- The energy stored in the capacitor for the above piezoelectric device would be CV^2 = (8.5nF)(2^2) = 34nJ.
