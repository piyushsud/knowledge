- Polarized caps are typically used when a very large capacitance is required, since the usage of an electrolyte allows them to have extremely high capacitances in a relatively small package. They are typically used to filter low-frequency noise. 

- low side injection is when the LO frequency is lower than the RF frequency, and high side injection is when the LO frequency is higher than the RF frequency. The mixer produces frequencies fRF - fLO and fRF + fLO, where fRF + fLO is typically filtered out for downconversion.
- In the case of high side injection, fRF - fLO is negative frequency. However, negative frequencies fold over the imaginary axis; for all practical purposes, it can be treated as a positive frequency. The sign of the frequency tells you if the wave is coming towards you or going away from you - however, this does not affect the frequency of the wave, which is a positive value. We use the negative sign as a convenient mathematical construct that has no meaning in reference to frequency. https://www.reddit.com/r/DSP/comments/iqldt/negative_frequencies/
