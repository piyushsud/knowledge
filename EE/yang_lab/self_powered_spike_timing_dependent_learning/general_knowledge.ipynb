Hebbian Learning Rule

- The hebbian learning rule states that if two connected neurons fire at the same time, the weight between them should increase.


- Long term potentiation and long term depression are phenomenon observed in biological neurons. Suppose that the time that a spike occurs one one end of a synapse is tpre and the time that occurs at the other end of a synapse is tpost. Then, long term potentation states that as tpre - tpost gets closer and closer to 0, the synaptic strength increases more and more. However, if tpre > tpost, that means that there was no causal relationship between the first spike and the second, so the synaptic strength decreases. This is known as long term depression. This makes sense, since there shouldn't be a link between two neurons if their firing is unrelated.

- The symmetric hebbian learning rule does not show a causal relationship, since it only depends on how correlated the two spikes are in time.


Piezoelectric Pressure Sensor 

- The piezoelectric effect is when a piezoelectric crystal is sandwiched between two metal plates (essentially a capacitor), it generates an electric charge when the crystal is compressed. Applying a constant force to a piezoelectric element will NOT cause additional charge to flow onto the plates of the capacitor - but changing the force will. Therefore, a piezoelectric device can be thought of as a force-controlled current source.
- Piezoelectric transducers produce an electric field that is proportional to the stress. 
- Stress is the force felt by an object per unit area, while pressure is the force applied to an object per unit area. They both have units of N/M^2.
- The voltage produced by a piezoelectric pressure sensor is given by:
- V = Ed = (g33F/A)*d,
- where F is the force felt by the sensor, A is the area of the object, and d is the distance between the parallel plates of the capacitor.
- The coefficient g is usually specified as g33, which is the induced electric field in direction 3 per unit stress applied in direction 3.
- The units of this equation make sense because g33 = Vm/N, F = N, A = m^2, and d = m, so g33F/A * d = V.
- Most piezoelectric devices have a very small capacitance, which means that any load that is less than a few Gohms will quickly discharge any charge generated by the piezoelectric device.
